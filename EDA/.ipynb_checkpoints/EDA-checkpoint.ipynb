## Categorizing movies into different cinema eras

### <font color = "pink"> *Cinema Eras* </font>

- **Silent Era (1902-1930)** : This period covers the early years of cinema before synchronized sound was introduced. It's marked by silent films and early experiments with film techniques.

- **Golden Age of Hollywood (1930-1960)** : This era includes the peak of the studio system, major film genres like musicals and film noir, and iconic stars. It's also the time of significant advancements in film technology and storytelling.

- **New Hollywood (1960-1980)** : Marked by the rise of innovative directors and films that challenged traditional norms. This era saw the emergence of independent films and new narrative styles.

- **Blockbuster Era (1980-1999)** : Defined by the rise of high-budget, high-grossing films and the dominance of blockbuster franchises. It's also a time of significant technological advancements like CGI.

- **Digital Age (2000-2018)** : The period where digital technology began to dominate filmmaking and distribution. This era includes the rise of digital effects, streaming services, and changes in audience consumption habits.

```python
# Defininig a function to categorize the eras

def categorize_era(year):
    if 1902 <= year <= 1929:
        return 'Silent Era (1902-1929)'
    elif 1930 <= year <= 1959:
        return 'Golden Age of Hollywood (1930-1959)'
    elif 1960 <= year <= 1979:
        return 'New Hollywood (1960-1979)'
    elif 1980 <= year <= 1999:
        return 'Blockbuster Era (1980-1999)'
    elif 2000 <= year <= 2009:
        return 'Digital Revolution Era (2000-2009)'
    elif 2010 <= year <= 2018:
        return 'Streaming Age (2010-2018)'
    else:
        return 'Unknown Era'
```

```python
# Apply the function to create the new column
final_df['releaseEra'] = final_df['releaseYear'].apply(categorize_era)
final_df.head()
```

```python
order = [
    'Silent Era (1902-1929)',
    'Golden Age of Hollywood (1930-1959)',
    'New Hollywood (1960-1979)',
    'Blockbuster Era (1980-1999)',
    'Digital Revolution Era (2000-2009)',
    'Streaming Age (2010-2018)'
]

# Convert 'releaseEra' to a categorical type with the defined order
final_df['releaseEra'] = pd.Categorical(
    final_df['releaseEra'], 
    categories=order, 
    ordered=True
)
```

### <font color = "pink"> **Count of ratings by release era** </font>

```python


# Filter out 'Unknown Era' if it exists
final_df_filtered = final_df[final_df['releaseEra'].isin(order)]

plt.figure(figsize=(20, 10))
sns.countplot(x='releaseEra', data=final_df_filtered, palette='viridis', order=order)

# Adjust the margins for more space between labels
plt.gca().margins(x=0.05)  # Adjust x margin

# Rotate the x-axis labels and adjust alignment
# plt.xticks(rotation=45, ha="right")
plt.xlabel('Year')
plt.ylabel('Count of Ratings')
plt.title('Count of Ratings by Rated Year')
plt.show()
```

# <font color='yellow'> EDA - Most/Highest Rated Movies and Genres </font>

### <font color = "pink"> **Most (not highest) rated movies in each era** </font>

#### Movies

```python
# Group by 'title' and calculate the number of ratings
grouped_df_2 = final_df.groupby(['releaseEra', 'title']).agg(
    num_ratings=('rating', 'size')
).reset_index()

# Remove movies without a legitimate era
filtered_df_2 = grouped_df_2[grouped_df_2['releaseEra'] != "Unknown Era"]

# Sort movies by average count from highest to lowest
sorted_df_2 = filtered_df_2.sort_values(by=['releaseEra', 'num_ratings'], ascending=False)

sorted_df_2
```

```python
# filtering out the top five in each category

sorted_df_2[sorted_df_2.groupby('releaseEra')['num_ratings'].rank('dense', ascending=False) <= 5]
```

### <font color = "pink"> **Highest rated movies in each era** </font>

#### Movies

```python
# Group by 'title' and calculate the number of ratings and average rating
grouped_df_3 = final_df.groupby(['releaseEra', 'title']).agg(
    num_ratings=('rating', 'size'),
    avg_rating=('rating', 'mean')
).reset_index()

# Filter movies with at least 30 ratings
filtered_df_3 = grouped_df_3[grouped_df_3['num_ratings'] >= 30]

# Sort movies by average rating from highest to lowest
sorted_df_3 = filtered_df_3.sort_values(by=['releaseEra', 'avg_rating'], ascending=False)

sorted_df_3
```

```python
# filtering out the top five in each category

sorted_df_3[sorted_df_3.groupby('releaseEra')['avg_rating'].rank('dense', ascending=False) <= 5]
```

# Creating a new dataframe to analyze genres

```python
# convert names series into string using str method 
# split the string on basis of comma delimiter 
# convert the series into list using to_list method 
# use stack to finally convert list elements to rows 
  
df_stack = pd.DataFrame(final_df.genres.str.split("|").to_list(), index=final_df.movieId).stack() 
df_stack = df_stack.reset_index(["movieId"]) 


df_stack.columns = ['movieId','genre']
  
print(df_stack) 
```

```python
# bringing in the rest of information related to each movie id

df_stack_full = df_stack.merge(final_df, how='inner', on="movieId")

df_stack_full.head()
```

### <font color = "pink"> **Most (not highest) rated movies in each genre** </font>

```python
# Group by 'title' and calculate the number of ratings

grouped_df_0 = (
    df_stack_full[['genre', 'userId', 'movieId','releaseEra']]
    .drop_duplicates()
    .groupby(['releaseEra','genre'])
    .agg(num_ratings=('movieId', 'size')) 
    .reset_index()
)

# Filter genres to not include useless categories
filtered_df_0 = grouped_df_0[(grouped_df_0['genre'] != "(no genres listed)")]
filtered_df_0_1 = filtered_df_0[(grouped_df_0['releaseEra'] != "Unknown Era")]

# Filter genres with at least 30 ratings
filtered_df_0_2 = filtered_df_0_1[filtered_df_0_1['num_ratings'] >= 30]


# Sort movies by average rating from highest to lowest
sorted_df_0 = filtered_df_0_2.sort_values(by=['releaseEra', 'num_ratings'], ascending=False)

sorted_df_0
```

```python
# filtering out the top five in each category

sorted_df_0[sorted_df_0.groupby('releaseEra')['num_ratings'].rank('dense', ascending=False) <= 5]
```

### <font color = "pink"> **Highest rated genres in each era** </font>

```python
# Group by 'title' and calculate the number of ratings

grouped_df_1 = (
    df_stack_full[['genre', 'userId', 'movieId','rating','releaseEra']]
    .drop_duplicates()
    .groupby(['releaseEra','genre'])
    .agg(num_ratings=('movieId', 'size'),
       avg_rating=('rating', 'mean')) 
    .reset_index()
)

# Filter movies with at least 30 ratings
filtered_df_1 = grouped_df_1[(grouped_df_1['genre'] != "(no genres listed)")]
filtered_df_1_1 = filtered_df_1[(grouped_df_1['releaseEra'] != "Unknown Era")]

# Filter genres with at least 30 ratings
filtered_df_1_2 = filtered_df_1_1[filtered_df_1_1['num_ratings'] >= 30]

# Sort movies by average rating from highest to lowest
sorted_df_1 = filtered_df_1_2.sort_values(by=['releaseEra', 'avg_rating'], ascending=False)

sorted_df_1
```

```python
# filtering out the top five in each category

sorted_df_1[sorted_df_1.groupby('releaseEra')['avg_rating'].rank('dense', ascending=False) <= 5]
```
